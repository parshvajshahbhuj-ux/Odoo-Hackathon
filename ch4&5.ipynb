```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
df
```

```python
print(df.info)
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=27) #underroot (total data) 768=27.... if any even number comes then +1 or -1 and take odd number
knn.fit(x_train, y_train)
pred=knn.predict(x_test)

#confusion metrices
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
cm=confusion_matrix(y_test,pred)
cm
#order[TN FP
#      FN TP]
# to find accuracy: (TP+TN)/(TP+FP+FN+TN) 

print("TP:",cm[1][1])
print("TN:",cm[0][0])
print("FN:",cm[1][0])
print("FP:",cm[0][1])
print("Accuracy:",accuracy_score(y_test,pred))
print("Sensitivity:",recall_score(y_test,pred))
```

Create a descision tree model for diabetes.csv dataset. display confusion matrix, accuracy score, error rate, specificity and sensitivity

```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
df
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,-1]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)

from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="entropy", max_depth=13)
m=model.fit(x_train,y_train)
print(model.get_depth())

from sklearn import tree
#print(tree.export_text(model))

y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Error Rate:", (1-(accuracy_score(y_test,y_pred))))
print("Sensitivity:",recall_score(y_test,y_pred))
#TN/(TN+FP)=specificity
print("Specificity:", cm[0][0]/(cm[0][0]+cm[0][1]))
print("Classification report \n", classification_report(y_test,y_pred))
```
