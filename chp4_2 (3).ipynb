```python
#create a decision treee

import pandas as pd

df=pd.read_csv("fish.csv")
df
```

```python
x=df.iloc[:,1:7]
y=df.iloc[:,0]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="entropy",max_depth=11)
m=model.fit(x_train,y_train)
print(model.get_depth())

y_pred=model.predict(x_test)

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Error Rate:", (1-(accuracy_score(y_test,y_pred))))
print("Classification report \n", classification_report(y_test,y_pred))
```

```python
import pandas as pd

df=pd.read_csv("diabetes.csv")
df
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)

from sklearn.svm import SVC
svm_model=SVC(kernel="rbf",random_state=42,C=10)
svm_model.fit(x_train,y_train)
y_pred=svm_model.predict(x_test)

from sklearn.metrics import confusion_matrix, accuracy_score, recall_score,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("TP:",cm[1][1])
print("TN:",cm[0][0])
print("FN:",cm[1][0])
print("FP:",cm[0][1])
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Sensitivity:",recall_score(y_test,y_pred))
print("classification_report",classification_report(y_test,y_pred))
```

```python
import pandas as pd
df=pd.read_csv("Employee_Attrition.csv")
df=df.drop("Employee_ID",axis=1)
df=pd.get_dummies(df,drop_first=True)

x=df.iloc[:,[0,1,2,3,4,6,7,8,9,10,11,12,13]]
y=df["Attrition"]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score
acc_list=[]

for k in range(1,26):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    acc_list.append(acc)
    print(f"k={k},Accuracy={acc}")
max_acc=max(acc_list)

best_k=(acc_list.index(max_acc))+1
print("------------------------------------------------------------------------------")
print("Best_k: ",best_k)

knn=KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x_train,y_train)

from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=150,random_state=42,max_depth=12,criterion="entropy")
m=model.fit(x_train,y_train)
y_pred1=m.predict(x_test)

acc=accuracy_score(y_test,y_pred1)
print("---------------------------------------------------------------------------------")
print("Accuracy:",acc)
cm=confusion_matrix(y_test,y_pred1)
print("---------------------------------------------------------------------------------")
print(cm)

import numpy as np
x_t=np.array([30,45000,5,3,10,1,0,0,1,0,1,0,0]).reshape(1,-1)
print("Prediction for knn:",knn.predict(x_t))
print("Prediction for rf:",model.predict(x_t))
```
