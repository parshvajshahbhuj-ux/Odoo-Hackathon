```python
import pandas as pd
df=pd.read_csv("Book1.csv")
df
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(y_test)
y_pred=(lr.predict(x_test))
print(" ")
print(y_pred)
```

```python
# to find m in y=mx+c
m=lr.coef_
m
```

```python
# to find c in y=mx+c
c=lr.intercept_
c
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("mae: ",mean_absolute_error(y_test,y_pred)) #1/n(sigma from 1 to n(actual value of x-predicted value of x))
print("mse: ",mean_squared_error(y_test,y_pred))#1/n(sigma from 1 to n((actual value of x-predicted value of x))^2)
print("r2score: ",r2_score(y_test,y_pred))#1-(sres/slot) = 1-er
```

```python
import pandas as pd
df=pd.read_csv("diabetes.csv")
df
```

```python
x=df[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]]
y=df["Outcome"]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=27)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,pred)
cm

print("IP:",cm[1][1])
print("IN:",cm[0][0])
print("FN:",cm[1][0])
print("FP:",cm[0][1])
print("accuracy",accuracy_score(y_test,pred))
# print("Sensitivity: ",recall_score(y_test,pred))
```

```python
import pandas as pd
df=pd.read_csv("Fish.csv")
df
```

```python
x=df[["Weight","Length1","Length2","Length3","Height","Width"]]
y=df["Species"]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=13)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,pred)
print(cm)
ac=accuracy_score(y_test,pred)

print("IP:",cm[1][1])
print("IN:",cm[0][0])
print("FN:",cm[1][0])
print("FP:",cm[0][1])
print("accuracy",ac)
print("error rate",(1-ac))
```

```python
# Split the dataset into features (X) and labels (y).
# Split the features and labels into training and testing sets with a test size of 0.2 and a random state of 42.
# Initialize a kNN classifier with 3 neighbors.
# Train the classifier on the training set.
# Make predictions on the test set.
# Calculate and print the accuracy score of the classifier.

import pandas as pd
df=pd.read_csv("Iris.csv")
df
```

```python
x=df[["Id","SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y=df["Species"]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,pred)
print(cm)
ac=accuracy_score(y_test,pred)

print("IP:",cm[1][1])
print("IN:",cm[0][0])
print("FN:",cm[1][0])
print("FP:",cm[0][1])
print("accuracy",ac)
print("error rate",(1-ac))
```
